```python
import sys
sys.path.insert(0, '/home/bekah/gPhoton2')
from pyarrow import parquet 
import pandas as pd 
from astropy.io import fits
import matplotlib.pyplot as plt
from pyarrow import parquet
from hostess.subutils import Viewer
from hostess.profilers import DEFAULT_PROFILER as PRO 
```

```python
# timing different run types of astrometry.net 

def time_run(name): 
    from hostess.profilers import DEFAULT_PROFILER as PRO 
    with PRO.context(name): 
        astrometry_hostess_run()
    return PRO.labels[name]['time']
```

```python
# get ra and dec 
eclipse = 815
og_aspect = parquet.read_table('/home/bekah/gPhoton2/gPhoton/aspect/aspect.parquet',
                                  filters=[('eclipse','==',eclipse)]).to_pandas()
rae = og_aspect.iloc[1]['ra']
dece = og_aspect.iloc[1]['dec']

xylist = "/media/bekah/BekahA/glcat/e00815/astrom/l2ts00330.xyls"
backplane = "/media/bekah/BekahA/glcat/e00815/e00815-nd-b02-f0001-t00330-g_dose.fits.gz"

# get width / height / pixels from image 

hdu = fits.open(backplane)

w = hdu[0].header['NAXIS1']
h = hdu[0].header['NAXIS2']
cx = w/2
cy = h/2


# run 1: normal run 
def astrometry_hostess_run(): 
    from hostess.subutils import Viewer
    solve_process = Viewer.from_command(
        "solve-field",
        xylist,
        overwrite=True,
        no_plots=True,
        dir_="/media/bekah/BekahA/glcat/e01532/astrom/",
        width=w,
        height=h,
        scale_units="arcsecperpix",
        scale_low=1.0,
        scale_high=1.6,
        N="none",
        U="none",
        B="none",
        M="none",
        R= "none",
        ra=rae,
        dec=dece,
        temp_axy=True,
        crpix_x=cx,
        crpix_y=cy)
    solve_process.wait()
    return solve_process.done
print("starting first run")
run1 = time_run("run1")
print("run1 done")
# run 2: uniformize 0
def astrometry_hostess_run(): 
    from hostess.subutils import Viewer
    solve_process = Viewer.from_command(
        "solve-field",
        xylist,
        overwrite=True,
        no_plots=True,
        dir_="/media/bekah/BekahA/glcat/e01532/astrom/",
        width=w,
        height=h,
        scale_units="arcsecperpix",
        scale_low=1.0,
        scale_high=1.6,
        N="none",
        U="none",
        B="none",
        M="none",
        R= "none",
        ra=rae,
        dec=dece,
        temp_axy=True,
        crpix_x=cx,
        crpix_y=cy,
        uniformize = 0)
    solve_process.wait()
    return solve_process.done

run2 = time_run("run2")
print("run2 done")

# run 3: uniformize 0
def astrometry_hostess_run(): 
    from hostess.subutils import Viewer
    solve_process = Viewer.from_command(
        "solve-field",
        xylist,
        overwrite=True,
        no_plots=True,
        dir_="/media/bekah/BekahA/glcat/e01532/astrom/",
        width=w,
        height=h,
        scale_units="arcsecperpix",
        scale_low=1.0,
        scale_high=1.6,
        N="none",
        U="none",
        B="none",
        M="none",
        R= "none",
        ra=rae,
        dec=dece,
        temp_axy=True,
        crpix_x=cx,
        crpix_y=cy,
        uniformize = 0)
    solve_process.wait()
    return solve_process.done

run3 = time_run("run3")
print("run3 done")

# run 4: no remove lines 
def astrometry_hostess_run(): 
    from hostess.subutils import Viewer
    solve_process = Viewer.from_command(
        "solve-field",
        xylist,
        overwrite=True,
        no_plots=True,
        dir_="/media/bekah/BekahA/glcat/e01532/astrom/",
        width=w,
        height=h,
        scale_units="arcsecperpix",
        scale_low=1.0,
        scale_high=1.6,
        N="none",
        U="none",
        B="none",
        M="none",
        R= "none",
        ra=rae,
        dec=dece,
        temp_axy=True,
        crpix_x=cx,
        crpix_y=cy,
        no_remove_lines=True)
    solve_process.wait()
    return solve_process.done

run4 = time_run("run4")
print("run4 done")

# run 5: uniformize = 0 AND remove lines 
def astrometry_hostess_run(): 
    from hostess.subutils import Viewer
    solve_process = Viewer.from_command(
        "solve-field",
        xylist,
        overwrite=True,
        no_plots=True,
        dir_="/media/bekah/BekahA/glcat/e01532/astrom/",
        width=w,
        height=h,
        scale_units="arcsecperpix",
        scale_low=1.0,
        scale_high=1.6,
        N="none",
        U="none",
        B="none",
        M="none",
        R= "none",
        ra=rae,
        dec=dece,
        temp_axy=True,
        crpix_x=cx,
        crpix_y=cy,
        no_remove_lines=True,
        uniformize = 0)
    solve_process.wait()
    return solve_process.done

run5 = time_run("run5")
print("run5 done")

# run 6: no temp axy
def astrometry_hostess_run(): 
    from hostess.subutils import Viewer
    solve_process = Viewer.from_command(
        "solve-field",
        xylist,
        overwrite=True,
        no_plots=True,
        dir_="/media/bekah/BekahA/glcat/e01532/astrom/",
        width=w,
        height=h,
        scale_units="arcsecperpix",
        scale_low=1.0,
        scale_high=1.6,
        N="none",
        U="none",
        B="none",
        M="none",
        R= "none",
        ra=rae,
        dec=dece,
        crpix_x=cx,
        crpix_y=cy)
    solve_process.wait()
    print(solve_process.err)
    return solve_process.done

run6 = time_run("run6")
print("run6 done")


# run 7: image
def astrometry_hostess_run(): 
    from hostess.subutils import Viewer
    solve_process = Viewer.from_command(
        "solve-field",
        backplane,
        overwrite=True,
        no_plots=True,
        dir_="/media/bekah/BekahA/glcat/e01532/astrom/",
        width=w,
        height=h,
        scale_units="arcsecperpix",
        scale_low=1.0,
        scale_high=1.6,
        N="none",
        U="none",
        B="none",
        M="none",
        R= "none",
        ra=rae,
        dec=dece,
        temp_axy=True,
        crpix_x=cx,
        crpix_y=cy)
    solve_process.wait()
    print(solve_process.err)
    return solve_process.done

run7 = time_run("run7")
print("run7 done")

# run 8: image w less inputs
def astrometry_hostess_run(): 
    from hostess.subutils import Viewer
    solve_process = Viewer.from_command(
        "solve-field",
        backplane,
        overwrite=True,
        no_plots=True,
        dir_="/media/bekah/BekahA/glcat/e01532/astrom/",
        scale_units="arcsecperpix",
        scale_low=1.0,
        scale_high=1.6,
        N="none",
        U="none",
        B="none",
        M="none",
        R= "none",
        ra=rae,
        dec=dece,
        temp_axy=True)
    solve_process.wait()
    print(solve_process.err)
    return solve_process.done

run8 = time_run("run8")
print("run8 done")
```
