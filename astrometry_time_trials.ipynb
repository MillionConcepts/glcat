```python
# timing different run types of astrometry.net 

def time_run(name, settings): 
    from hostess.profilers import DEFAULT_PROFILER as PRO 
    with PRO.context(name): 
        refine_eclipse(
               eclipse=settings['eclipse'],
               xylist=settings['xylist'],
               band="NUV",
               aspect_root="/media/bekah/BekahA/glcat",
               gphoton_root="/home/bekah/gPhoton2",
               ext="gzip",
               threads=settings['threads'])
    return PRO.labels[name]['time']
```

```python
# get ra and dec 
eclipse = 815
og_aspect = parquet.read_table('/home/bekah/gPhoton2/gPhoton/aspect/aspect.parquet',
                                  filters=[('eclipse','==',eclipse)]).to_pandas()
rae = og_aspect.iloc[1]['ra']
dece = og_aspect.iloc[1]['dec']

xylist = "/media/bekah/BekahA/glcat/e00815/astrom/l2ts00330.xyls"
backplane = "/media/bekah/BekahA/glcat/e00815/e00815-nd-b02-f0001-t00330-g_dose.fits.gz"
adir = "/media/bekah/BekahA/glcat/e00815/astrom/"
# get width / height / pixels from image 

hdu = fits.open(backplane)

w = hdu[0].header['NAXIS1']
h = hdu[0].header['NAXIS2']
cx = w/2
cy = h/2
```

```python


# run 8: image w less inputs
def astrometry_hostess_run(): 
    from hostess.subutils import Viewer
    solve_process = Viewer.from_command(
        "solve-field",
        backplane,
        overwrite=True,
        no_plots=True,
        dir_=adir,
        scale_units="arcsecperpix",
        scale_low=1.0,
        scale_high=1.6,
        N="none",
        U="none",
        B="none",
        M="none",
        R= "none",
        ra=rae,
        dec=dece,
        temp_axy=True)
    solve_process.wait()
    print(solve_process.err)
    return solve_process.done

run8 = time_run("run8")
print("run8 done")

PRO.labels
```
